```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "C:\infosys_springboard_internship",  
    image_size=(224, 224),    
    batch_size=32,            
    shuffle=True               
)

class_names = dataset.class_names

def display_random_images(dataset, class_names, num_images=3):
    for class_index, class_name in enumerate(class_names):

        images, labels = next(iter(dataset.unbatch().filter(lambda x, y: y == class_index).batch(num_images)))
       
        plt.figure(figsize=(15, 5))
        for i in range(num_images):
            plt.subplot(1, num_images, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.title(class_name)
            plt.axis('off')
        plt.show()
display_random_images(dataset, class_names)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Load the dataset
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "C:\infosys_springboard_internship" ,
    image_size=(224, 224),
    batch_size=32,
    shuffle=True
)

data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal_and_vertical"),
    tf.keras.layers.RandomRotation(0.2),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1),
])

def display_augmented_images(dataset, num_images=5):
    plt.figure(figsize=(15, 5))
    for i in range(num_images):
    
        image, label = next(iter(dataset))
        augmented_image = data_augmentation(image)
        
        plt.subplot(1, num_images, i + 1)
        plt.imshow(augmented_image[0].numpy().astype("uint8")) 
        plt.axis('off')
    plt.show()

display_augmented_images(dataset)
```
